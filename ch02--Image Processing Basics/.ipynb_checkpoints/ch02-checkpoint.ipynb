# 第二章：影像處理基礎

```python
# 此自訂函數可透過 matplotlib 顯現影像
import numpy as np
import cv2
import matplotlib.pyplot as plt

def showImage(title, img):
    ishow = cv2.cvtColor(img, cv2.COLOR_BGR2RGB) # for plt.imshow()
    plt.imshow(ishow)
    plt.axis('off')
    plt.title(title)
    plt.show()
```

```python
# 例 2.1: 利用 numpy 的 zeros() 函式產生二值化黑白影像
import cv2
import numpy as np

img = np.zeros((8,8), dtype = np.uint8) # uint8: unsigned 8-bit integer
print("img =\n", img)
cv2.imshow("one", img)
showImage("one", img)
print("讀取像素點 img[0,3] =", img[0,3])

img[0,3] = 255
print("修改後 img =\n",img)
print("讀取修改後像素點 img[0,3] =", img[0,3])
cv2.imshow("two", img)
showImage("two", img)

cv2.waitKey()
cv2.destroyAllWindows()
```

```python
# 例 2.2: 讀取灰階影像，並進行修改
import cv2

img = cv2.imread("lena.bmp", 0) # 0: cv2.IMREAD_GRAYSCALE (灰階影像)
cv2.imshow("before", img)
showImage("before", img)

for i in range(10,100):       # i: row
    for j in range(80, 100):  # j: column
        img[i,j] = 255        # 設成白色
cv2.imshow("after",img)
showImage("after", img)

cv2.waitKey()
cv2.destroyAllWindows()
```

```python
# 例 2.3: 利用 numpy 產生藍綠紅三個通道（三維矩陣）
import numpy as np
import cv2

#-----------藍色通道（blue channel）--------------
blue = np.zeros((300,300,3), dtype = np.uint8)
blue[:,:,0] = 255
print("blue=\n", blue)
cv2.imshow("blue", blue)
showImage("blue", blue)

#-----------綠色通道（green channel）--------------
green = np.zeros((300,300,3), dtype = np.uint8)
green[:,:,1] = 255
print("green=\n", green)
cv2.imshow("green", green)
showImage("green", green)

#-----------紅色通道（red channel）----------------
red = np.zeros((300,300,3), dtype = np.uint8)
red[:,:,2] = 255
print("red=\n", red)
cv2.imshow("red", red)
showImage("red", red)

#-----------關閉視窗（release window）--------------
cv2.waitKey()
cv2.destroyAllWindows()
```

```python
# 例 2.4: 利用 numpy 產生一個三維矩陣，觀察藍綠紅三個通道變化
import numpy as np
import cv2

img = np.zeros((300,300,3), dtype=np.uint8)
img[:,  0:100,0] = 255  # columns   0~ 99: channel 0 - Blue
img[:,100:200,1] = 255  # columns 100~199: channel 1 - Green
img[:,200:300,2] = 255  # columns 200~299: channel 2 - Red
print("img = \n", img)
cv2.imshow("img", img)
showImage("img", img)

cv2.waitKey()
cv2.destroyAllWindows()
```

```python
# 例 2.5: 利用 numpy 產生三維矩陣，模擬 RGB 影像，並進行走訪與修改
import numpy as np

img = np.zeros((2,4,3), dtype = np.uint8)  # 產生 2x4x3 的矩陣 (2 rows, 4 cols, 3 channels)
print("img =\n", img)
print("讀取像素點 img[0,3] =", img[0,3])
print("讀取像素點 img[1,2,2] =", img[1,2,2])
showImage("before", img)

# 修改影像
img[0,3] = 255
img[0,0] = [66,77,88]
img[1,1,1] = 3
img[1,2,2] = 4
img[0,2,0] = 5
print("修改後img\n", img)
print("讀取修改後像素點 img[1,2,2] =", img[1,2,2])
showImage("after", img)
```

```python
# 例 2.6: 讀取彩色影像，直接對其進行走訪與修改
import cv2

img = cv2.imread("lenacolor.png")
cv2.imshow("before", img)
showImage("before", img)

# ---走訪---
print("访问img[0,0]=", img[0,0])
print("访问img[0,0,0]=", img[0,0,0])
print("访问img[0,0,1]=", img[0,0,1])
print("访问img[0,0,2]=", img[0,0,2])
print("访问img[50,0]=", img[50,0])
print("访问img[100,0]=", img[100,0])

# ---修改---
# 区域1
for i in range(0,50):
    for j in range(0,100):
        for k in range(0,3):
            img[i,j,k]=255  #白色

# 区域2
for i in range(50,100):
    for j in range(0,100):
        img[i,j]=[128,128,128]  #灰色

# 区域3            
for i in range(100,150):
    for j in range(0,100):
        img[i,j]=0          #黑色

cv2.imshow("after",img)
showImage("after", img)

# ---修改後，再次訪問---
print("修改后img[0,0]=", img[0,0])
print("修改后img[0,0,0]=", img[0,0,0])
print("修改后img[0,0,1]=", img[0,0,1])
print("修改后img[0,0,2]=", img[0,0,2])
print("修改后img[50,0]=", img[50,0])
print("修改后img[100,0]=", img[100,0])

cv2.waitKey()
cv2.destroyAllWindows()
```

```python
# 例 2.7: 使用 numpy 產生一個二維隨機陣列，模擬灰階影像，走訪並修改
import numpy as np

img = np.random.randint(10, 99, size = [5,5], dtype = np.uint8) # 5x5 矩陣，隨機值介於10~98
print("img=\n", img)
print("读取像素点img.item(3,2)=", img.item(3,2))
showImage("before", img)

img.itemset((3,2), 255)  # 修改
print("修改后img=\n", img)
print("修改后像素点img.item(3,2)=", img.item(3,2))
showImage("after", img)
```

```python
# 例 2.8: 產生一個灰階影像，其像素值為隨機數
import numpy as np
import cv2

img = np.random.randint(0, 256, size = [512,512], dtype = np.uint8) # 512x512 矩陣，其隨機值介於0~255
cv2.imshow("demo",img)
showImage("demo", img)

cv2.waitKey()
cv2.destroyAllWindows()
```

```python
# 例 2.9: 讀取灰階影像，走訪及修改
import cv2

img = cv2.imread("lena.bmp", 0)

# 测试读取、修改单个像素值
print("读取像素点img.item(3,2)=", img.item(3,2))
img.itemset((3,2),255)
print("修改后像素点img.item(3,2)=", img.item(3,2))

# 测试修改一个区域的像素值
cv2.imshow("before",img)
showImage("before", img)

for i in range(10,100):
    for j in range(80,100):
        img.itemset((i,j),255)
cv2.imshow("after",img)
showImage("after", img)

cv2.waitKey()
cv2.destroyAllWindows()
```

```python
# 例 2.10: 使用 numpy 產生一個三維隨機矩陣，模擬 RGB 彩色影像，走訪並修改
import numpy as np

# ---隨機產生---
img = np.random.randint(10, 99, size = [2,4,3], dtype = np.uint8)
print("img=\n",img)
showImage("before", img)

# ---走訪---
print("读取像素点img[1,2,0]=", img.item(1,2,0))
print("读取像素点img[0,2,1]=", img.item(0,2,1))
print("读取像素点img[1,0,2]=", img.item(1,0,2))

# ---修改---
img.itemset((1,2,0), 255)
img.itemset((0,2,1), 255)
img.itemset((1,0,2), 255)

#---再次走訪---
print("修改后img=\n",img)
print("修改后像素点img[1,2,0]=",img.item(1,2,0))
print("修改后像素点img[0,2,1]=",img.item(0,2,1))
print("修改后像素点img[1,0,2]=",img.item(1,0,2))
showImage("after", img)
```

```python
# 例 2.11: 產生一幅彩色影像，其像數值為隨機
import cv2
import numpy as np

img = np.random.randint(0, 256, size = [256,256,3], dtype = np.uint8) # 256x256x3 矩陣，其值介於0~255之間
cv2.imshow("demo",img)
showImage("demo", img)

cv2.waitKey()
cv2.destroyAllWindows()
```

```python
# 例 2.12: 讀取彩色影像，走訪與修改
import cv2

img = cv2.imread("lenacolor.png")
cv2.imshow("before",img)
showImage("before", img)

# ---走訪---
print("访问img.item(0,0,0)=", img.item(0,0,0))
print("访问img.item(0,0,1)=", img.item(0,0,1))
print("访问img.item(0,0,2)=", img.item(0,0,2))

# ---修改---
for i in range(0,50):
    for j in range(0,100):
        for k in range(0,3):
            img.itemset((i,j,k), 255)     #白色

# ---再次走訪---
cv2.imshow("after",img)
showImage("after", img)
print("修改后img.item(0,0,0)=", img.item(0,0,0))
print("修改后img.item(0,0,1)=", img.item(0,0,1))
print("修改后img.item(0,0,2)=", img.item(0,0,2))

cv2.waitKey()
cv2.destroyAllWindows()
```

```python
# 例 2.13: 讀取 lena 臉部訊息（ROI），並將其顯示出來
import cv2

a = cv2.imread("lenacolor.png", cv2.IMREAD_UNCHANGED)

# Region of Interest (ROI)
face = a[220:400, 250:350] # 取出 rows 220~399, cols 250~349

cv2.imshow("original", a)
showImage("original", a)

cv2.imshow("face", face)
showImage("face", face)

cv2.waitKey()
cv2.destroyAllWindows()
```

```python
# 例 2.14: 對 lena 影像的臉部進行打碼
import cv2
import numpy as np

a = cv2.imread("lenacolor.png", cv2.IMREAD_UNCHANGED)
cv2.imshow("original", a)
showImage("original", a)

face = np.random.randint(0, 256, (180,100,3)) # 180x100x3 矩陣，隨機值介於 0~255

# ---打碼（貼上以上隨機圖片）---
a[220:400, 250:350] = face
cv2.imshow("result",a)
showImage("result", a)

cv2.waitKey()
cv2.destroyAllWindows()
```

```python
# 例 2.15: 將 ROI 複製到另一幅影像
import cv2

lena = cv2.imread("lena512.bmp", cv2.IMREAD_UNCHANGED)
dollar = cv2.imread("dollar.bmp", cv2.IMREAD_UNCHANGED)
cv2.imshow("lena", lena)
showImage("lena", lena)

cv2.imshow("dollar", dollar)
showImage("dollar", dollar)

# ---從lena影像取出 ROI---
face = lena[220:400, 250:350]

# ---貼到另一幅影像上---
dollar[160:340, 200:300] = face
cv2.imshow("result",dollar)
showImage("result", dollar)

cv2.waitKey()
cv2.destroyAllWindows()
```

```python
# 例 2.16: 通道拆分
import cv2

lena = cv2.imread("lenacolor.png")
cv2.imshow("lena1",lena)

# ---取出 B,G,R 通道（channels）---
b=lena[:,:,0]
g=lena[:,:,1]
r=lena[:,:,2]

# ---顯示個別通道---
cv2.imshow("b", b)
showImage("b", b)

cv2.imshow("g", g)
showImage("g", g)

cv2.imshow("r", r)
showImage("r", r)

# ---設定 lena 通道0（藍色）為 0---
lena[:,:,0]=0
cv2.imshow("lenab0",lena)
showImage("lenab0", lena)

# ---再設定 lena 通道1（綠色）為 0，剩下通道2（紅色）---
lena[:,:,1]=0
cv2.imshow("lenab0g0",lena)
showImage("lenab0g0", lena)

cv2.waitKey()
cv2.destroyAllWindows()
```

```python
# 例 2.17: 使用 cv2.split() 拆分通道
import cv2

lena = cv2.imread("lenacolor.png")

# ---使用 cv2.split() 拆分通道---
b, g, r = cv2.split(lena)

# ---顯示個別通道---
cv2.imshow("B",b)
showImage("B", b)

cv2.imshow("G",g)
showImage("G", g)

cv2.imshow("R",r)
showImage("R", r)

cv2.waitKey()
cv2.destroyAllWindows()
```

```python
# 例 2.18: 使用 cv2.merge() 合併通道
import cv2

lena = cv2.imread("lenacolor.png")

# ---使用 cv2.split() 拆分通道---
b, g, r = cv2.split(lena)

# ---使用 cv2.merge() 合併通道，順序為 B,G,R ---（正確合併成原圖）
bgr = cv2.merge([b,g,r])

# ---使用 cv2.merge() 合併通道，順序為 R,G,B ---（錯誤）
rgb = cv2.merge([r,g,b])

# ---顯示合併結果---
cv2.imshow("lena",lena)
showImage("lena", lena)

cv2.imshow("bgr",bgr)
showImage("bgr", bgr)

cv2.imshow("rgb",rgb)
showImage("rgb", rgb)

cv2.waitKey()
cv2.destroyAllWindows()
```

```python
# 例 2.19: 影像常用屬性
import cv2

gray = cv2.imread("lena.bmp", 0)
color = cv2.imread("lenacolor.png")
showImage("gray", gray)
showImage("color", color)

print("图像gray属性：")
print("gray.shape=", gray.shape) # (rows, cols)
print("gray.size=", gray.size)   # 像素數目 rows x cols
print("gray.dtype=", gray.dtype) # 像素資料型態
print()

print("图像color属性：")
print("color.shape=", color.shape) # (rows, cols, channels)
print("color.size=", color.size)   # rows x cols x channels
print("color.dtype=", color.dtype) # 像素資料型態
```
